```python
import pandas as pd
import matplotlib.pyplot as plt
import os

class DataAnalyzer:
    def __init__(self, files):
        self.files = files
        
    def analyze_column(self, series, col_name, file_name):
        series = series.dropna()
        print(f"\n--- {file_name} | Column: {col_name} ---")
        
        if pd.api.types.is_numeric_dtype(series):
            self._analyze_numeric(series)
        else:
            self._analyze_categorical(series)
            
        self._plot_value_counts(series, col_name, file_name)
        
    def _analyze_numeric(self, series):
        print(f"Type: numeric")
        print(f"Mean: {series.mean():.4f}")
        print(f"Median: {series.median():.4f}")
        
    def _analyze_categorical(self, series):
        print(f"Type: categorical")
        
    def _plot_value_counts(self, series, col_name, file_name):
        vc = series.value_counts()
        
        if vc.empty or vc.iloc[0] == 1:
            print("Все значения уникальны, моды нет")
            return
            
        modes = series.mode().tolist()
        print(f"Mode(s): {modes}")
        
        plt.figure(figsize=(10, 5))
        vc.head(10).plot(kind='bar')
        plt.title(f"Top 10 — {file_name} | {col_name}")
        plt.ylabel("Count")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
        
    def analyze_all_files(self):
        """Запуск анализа для всех файлов"""
        for file_path in self.files:
            try:
                df = pd.read_excel(file_path)
                file_name = os.path.basename(file_path)
                print(f"\n{'='*30}\nAnalyzing file: {file_name}\n{'='*30}")
                
                for col in df.columns:
                    self.analyze_column(df[col], col, file_name)
                    
            except Exception as e:
                print(f"\nError processing {file_path}: {str(e)}")

# Пример использования
if __name__ == "__main__":
    files = [
        'Data_sets_&_Data_set_generation/Data_sets/dataset_5000.xlsx'
        # 'Data_sets_&_Data_set_generation/Data_sets/dataset_50000.xlsx',
        # 'Data_sets_&_Data_set_generation/Data_sets/dataset_120000.xlsx',
    ]
    
    analyzer = DataAnalyzer(files)
    analyzer.analyze_all_files()
```
